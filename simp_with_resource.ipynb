```python
fec_pars = {'A': {'OP50': [22.65, 68.45, 57.05, 33.4,  4.97], 'Novo': [11.66, 62.53, 47.13, 13.94,  0.72]},
            'C': {'OP50': [19.8 , 60.3 , 43.02, 19.9,  6.6], 'Novo': [16.88, 80.77, 77.7 , 16.28,  1.4]}}
```

```python
def generate_matrix(r, r_min=100):
    U = np.identity(10) * 0
    
    γ_JE  = lambda r: 0.04 if r > r_min else 0.0
    γ_DJ = lambda r: 0.04 if r < r_min else 0.0
    γ_YD = lambda r: 0.04 if r > r_min else 0.0
    γ_YJ = lambda r: 0.04 if r > r_min else 0.0
    γ_RAY = 0.08
    γ_RR = 0.031
    γ_OR = 0.031
    δ_E = lambda r: 0.99 if r > r_min else 0.9
    δ_J = lambda r: 0.99 if r > r_min else 0.9
    δ_Y = lambda r: 0.99 if r > r_min else 0.9
    δ_RA = lambda r: 0.99 if r > r_min else 0.9
    δ_OA = lambda r: 0.99 if r > r_min else 0.9
    δ_D = 1
    U[0][0] = δ_E(r) * (1 - γ_JE(r))
    # juvenile ra(r)tes
    U[1][0] = δ_E(r) * γ_JE(r)
    U[1][1] = δ_J(r) * (1 - γ_YJ(r)) * (1 - γ_DJ(r))
    # dauer rates
    U[2][1] = δ_J(r) * γ_DJ(r)
    U[2][2] = δ_D * (1 - γ_YD(r))
    # young adults
    U[3][1] = δ_J(r) * γ_YJ(r)
    U[3][2] = δ_D * γ_YD(r)
    U[3][3] = δ_Y(r) * (1 - γ_RAY)
    # reproducing adults
    U[4][3] = δ_Y(r) * γ_RAY
    U[4][4] = δ_RA(r) * (1 - γ_RR)

    U[5][4] = δ_RA(r) * γ_RR
    U[5][5] = δ_RA(r) * (1 - γ_RR)

    U[6][5] = δ_RA(r) * γ_RR
    U[6][6] = δ_RA(r) * (1 - γ_RR)

    U[7][6] = δ_RA(r) * γ_RR
    U[7][7] = δ_RA(r) * (1 - γ_RR)

    U[8][7] = δ_RA(r) * γ_RR
    U[8][8] = δ_RA(r) * (1 - γ_OR)

    # old adults
    U[9][8] = δ_RA(r) * γ_OR
    U[9][9] = δ_OA(r)
    
    return U
```

```python
def gen_fec_matrix(strain, food_type, rate):
    """Generate fecundity matrix.

    Args:
        strain (str): A (RSC017) or C (RS5405)
        food_type (str): The bacterial diet (Novo or OP50)

    Returns:
        numpy matrix: fecundity values for each breeding developmental stage.
    """
    F = np.identity(10) * 0
    count = 0
    for i in np.arange(4, 9, 1):
        F[0][i] = rate*fec_pars[strain][food_type][count]
        count += 1
    return F
```

```python
n0 = np.array([[0], [50], [0], [0], [0], [0], [0], [0], [0], [0]])
s0 = 1e4
c = 0.05
n_list = []
for i in range(150):
    t_mat = generate_matrix(s0)
    n0 = np.array(np.matmul(t_mat + f_mat, n0))
    s0 -= c*n0[1][0] + np.sum(n0[3:])
    if s0 < 0:
        s0=0
    n_list.append(n0)
```
