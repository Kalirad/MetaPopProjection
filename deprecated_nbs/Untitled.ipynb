```python
def simp_consum_resource(n0, r, θ, c, r_growth, slop, r_min, t):
    γ_JE  = 1
    γ_RAJ = lambda r: np.divide(0.5, 1 + np.exp(-1*slop*(r - r_min), dtype=np.float128))
    γ_DJ = lambda r: np.divide(0.5, 1 + np.exp(1*slop*(r - r_min), dtype=np.float128))
    γ_OR = 0.15
    γ_RAD = lambda r: np.divide(0.1, 1 + np.exp(-1*slop*(r - r_min), dtype=np.float128))
    δ_E = lambda r: np.divide(0.99, 1 + np.exp(-1*slop*(r - r_min), dtype=np.float128))
    δ_J = lambda r: np.divide(0.99, 1 + np.exp(-1*slop*(r - r_min), dtype=np.float128))
    δ_RA = lambda r: np.divide(0.99, 1 + np.exp(-1*slop*(r - r_min), dtype=np.float128))
    δ_OA = lambda r: np.divide(0.99, 1 + np.exp(-1*slop*(r - r_min), dtype=np.float128))
    δ_D = 0.99
    F = np.identity(5) * 0
    F[0][3] =γ_OR*180
    n_list = []
    r_list = []
    for i in range(t):
        U_simp = np.matrix([[δ_E(r)*(1 - γ_JE),                          0,               0,               0,    0],
                            [      δ_E(r)*γ_JE, δ_J(r)*(1 - γ_RAJ(r))*(1 - γ_DJ(r)),               0,               0,    0],
                            [             0,                   δ_J(r)*γ_DJ(r), δ_D*(1 - γ_RAD(r)),               0,    0],
                            [             0,                  δ_J(r)*γ_RAJ(r),       δ_D*γ_RAD(r), δ_RA(r)*(1 - γ_OR),    0],
                            [             0,                          0,               0,       δ_RA(r)*γ_OR, δ_OA(r)]])
        n0 = np.array(np.matmul(U_simp + F, n0))
        r = (r_growth -  c*np.sum(n0[1] + n0[3:]))*r + θ
        if r < 0:
            r = 0
        n_list.append(n0)
        r_list.append(r)
    return n_list, r_list
```

```python
n0 = np.array([[0], [20], [0], [0], [0]])
r0 = 100
r_growth = 2
θ = 1
c = 0.01
r_min = 50
test = simp_consum_resource(n0, r0, θ, c, r_growth, 0.2, r_min, 100)
```
